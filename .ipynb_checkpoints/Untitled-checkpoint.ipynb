```python
import lattice_functions as lf
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np
from scipy.constants import Boltzmann as kB
```

```python
def draw_lattice(lattice):
    im = plt.imshow(lattice,cmap='magma')
    return [im]
```

```python
def init_cluster(full_lat):
    cluster = random_position(len(full_lat))
    return cluster

def random_position(L):
    x = np.random.randint(low=0,high=L)
    y = np.random.randint(low=0,high=L)
    return (x,y)
```

```python
def wolff(T,lat,J=1):
    threeshold = 1 - np.exp(-2*J/T)
    cluster = np.copy(lat)
    make_cluster(lat,cluster,threeshold)
    lat = flip_cluster(lat,cluster)
    return [draw_lattice(lat)]
```

```python
def make_cluster(lat,cluster,threeshold):
    random_spin = random_position(len(lat))
    cluster[random_spin] = 0
    cluster=check_neighbours(lat,cluster,random_spin,threeshold)
    return cluster
```

```python
def flip_coin():
    return np.random.rand()
```

```python
def flip_cluster(lat,cluster):
    lat[np.where(cluster == 0)] *= -1
    return lat
```

```python
def check_neighbours(lat,cluster,spin,threeshold):  
    
    nn = np.array([-1,1])  
    
    for i in nn: 
        
        neighbors = [tuple([(x+y)%len(lat) for (x,y) in zip(spin,(i,0))]),tuple([(x+y)%len(lat) for (x,y) in zip(spin,(0,i))])]
        
        for neighbor in neighbors:
                
            if lat[spin] == lat[neighbor] and cluster[neighbor] != 0:
                
                if flip_coin() > threeshold:
                    
                    cluster[neighbor] = 0 
                    check_neighbours(lat,cluster,neighbor,threeshold)
```

```python
# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)
```
